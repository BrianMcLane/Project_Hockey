```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import numpy as np
import sklearn
```

```python
data = 'data_for_machine.csv'
```

```python
data_df = pd.read_csv(data, encoding='utf-8')
```

```python
data_df.head()
```

```python
data_df.columns
```

```python
target = data_df['decision']
target_name = ['negative', 'positive']
```

```python
info = data_df.drop('decision', axis=1)
feature_names = info.columns
info.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(info, target, random_state=42)
```

```python
from sklearn import tree
```

```python
clf = tree.DecisionTreeClassifier()
clf = clf.fit(X_train, y_train)
clf.score(X_test, y_test)
```

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(n_estimators=200)
rf = rf.fit(X_train, y_train)
rf.score(X_test, y_test)
```

```python
sorted(zip(rf.feature_importances_, feature_names), reverse=True)
```
