```python
import pandas as pd
import numpy as np
```

```python
game_skater_stats = "hockey_data/game_skater_stats.csv"
game_goalie_stats = "hockey_data/game_goalie_stats.csv"
player_info = "hockey_data/player_info.csv"
```

```python
# Import the books.csv file as a DataFrame
game_skater_stats_df = pd.read_csv(game_skater_stats, encoding="utf-8")
game_goalie_stats_df = pd.read_csv(game_goalie_stats)
player_info_df = pd.read_csv(player_info)
```

```python
game_skater_stats_df = game_skater_stats_df.loc[game_skater_stats_df['team_id'] == 16]
game_skater_stats_df
```

```python
game_goalie_stats_df = game_goalie_stats_df.loc[game_goalie_stats_df['team_id'] == 16]
game_goalie_stats_df
```

```python
player_list = game_skater_stats_df["player_id"].unique()
player_list = list(player_list)
len(player_list)
```

```python
goalie_id = game_goalie_stats_df["player_id"].unique()
goalie_id = list(goalie_id)
len(goalie_id)
```

```python
player_list.extend(goalie_id)
len(player_list)
```

```python
boolean_series = player_info_df.player_id.isin(player_list) 

filtered_df = player_info_df[boolean_series]
filtered_df
```

```python
u = filtered_df.select_dtypes(object)
filtered_df[u.columns] = u.apply(
    lambda x: x.str.encode('ascii', 'ignore').str.decode('ascii'))
filtered_df.sample(100)
```

```python
birthCity_list = list(filtered_df['birthCity'])
birthCity_list
```

```python
filtered_df.to_csv("clean_player_info.csv")
```
