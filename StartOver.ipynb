```python
import pandas as pd
import numpy as np
```

```python
game_skater_stats = "hockey_data/game_skater_stats.csv"
game_goalie_stats = "hockey_data/game_goalie_stats.csv"
player_info = "hockey_data/player_info.csv"
```

```python
game_skater_stats_df = pd.read_csv(game_skater_stats, encoding="utf-8")
game_goalie_stats_df = pd.read_csv(game_goalie_stats)
player_info_df = pd.read_csv(player_info)
```

```python
game_skater_stats_df.dropna()
```

```python
playerPos_df = player_info_df[['player_id', 'primaryPosition']]
playerPos_df
```

```python
merged_df = game_skater_stats_df.merge(playerPos_df, how="inner", on="player_id")
merged_df.drop_duplicates(inplace=True)
merged_df
```

```python
game = "hockey_data/game.csv"
game_df = pd.read_csv(game, encoding="utf-8")
```

```python
game_df = game_df[['game_id', 'season', 'date_time_GMT']].drop_duplicates()
```

```python
merged2 = merged_df.merge(game_df, how="inner", on="game_id")
merged2
```

```python
groupTest = merged2.groupby(['game_id', 'team_id', 'primaryPosition'])[['timeOnIce', 'assists', 'goals', 'shots', 'hits', 'powerPlayGoals', 'powerPlayAssists', 'penaltyMinutes', 'faceOffWins', 'takeaways', 'giveaways', 'shortHandedGoals', 'shortHandedAssists', 'blocked', 'evenTimeOnIce', 'shortHandedTimeOnIce', 'powerPlayTimeOnIce']].sum()
groupTest
```
